```python
import random
import os
import glob
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
import lorem
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
```

```python
# Get a list of nouns, this will be the basis of our Titles
noun_list = []
for synset in wn.all_synsets('n'):
    for lemma in synset.lemmas():
        noun_list.append(lemma.name())

# A collection of first and last names Ihave collected
result = pd.read_csv('Random_names_master.csv')
first_name_list = result['first_name'].tolist()
last_name_list = result['last_name'].tolist()
```

```python
# todo: add random number and random amount for price
class DataGen:
    """
    """
    def __init__(self,size):
        self._size = size
        
    def list_pick(self,choice_list,uniform = True,probability_array = None):
        """list_pick() -> list
          Keyword arguments:
          choice_list       -- list of items you want to choose from
          uniform           -- boolean parameter for uniform or non uniform selection from your list (default = True)
          probability_array -- if uniform = False this array will determin the probability you want each list element 
                               to be selected with (default = None)
                               IMPORTANT: probability_array elements must add up to 1.
        """
        result_list = []
        if uniform == True:
            for i in range(self._size):
                result_list.append(choice_list[random.randint(0,len(choice_list)-1)])
            return result_list
        elif uniform == False and probability_array == None:
            print('Please choose a probability array for your non uniform selection')
            pass
        elif uniform == False and probability_array != None:
            _choice_list = choice_list
            _probability_array = probability_array
            result_list = np.random.choice(_choice_list, self._size, p=_probability_array).tolist()
            return result_list
        else:
            print('Error: Something wrong! Please read the docstring for guidance on each parameter')
            pass
        
    def descr_gen(self): 
        """descr_gen() -> list
        """
        result_list = []
        for i in range(self._size):
            result_list.append(lorem.sentence())
        return result_list
    
    def title_gen(self,list_of_words = noun_list):
        """title_gen() -> list
            Keyword arguments:
            list_of_words -- list of words you want to choose from
        """
        result_list = []
        for i in range(self._size):
            result_list.append(list_of_words[random.randint(0,len(list_of_words)-1)].lstrip() +" "+ list_of_words[random.randint(0,len(list_of_words)-1)])
        return result_list
    
    def date_gen(self,start_year= 1950, end_year= 2020,set_month=None):
        """date_gen() -> list
        This function will return a random date between two years.
           Keyword arguments:
           start_year -- start limit of your year 
           end_year -- end limit of your year
           set_month -- if you want to hard code the month
        """
        result_list = []
        for i in range(self._size):
            year = random.choice(range(start_year, end_year))
            if set_month == None:
                month = random.choice(range(1, 13))
            else:
                month = set_month
            day = random.choice(range(1, 29))
            result_list.append(f'{year}/{month}/{day}') # datetime(year, month, day)
        return result_list
    
    def gen_random_name(self,list_1 = first_name_list,list_2 = last_name_list):
        """gen_random_name() -> list
            Keyword arguments:
            list_1 -- first list you want to randomly pick from
            list_2 -- second list you want to randomly pick from
        """        
        result_list = []
        for i in range(self._size):
            end_range_f = len(list_1) - 1
            end_range_l = len(list_2) - 1
            name = str(list_1[random.randint(0,end_range_f)].rstrip().lstrip())+" "+str(list_2[random.randint(0,end_range_l)])
            result_list.append(name)
        return result_list
```

## Example of uses

Let's say we need to mock up data for a publishing licensing company


```python
# Generate data on books
Books = DataGen(10)
book_data = pd.DataFrame({
    'author_name': Books.gen_random_name(),
    'publishing_ate': Books.date_gen(2000,2018),
    'title':Books.title_gen(),
    'descr' : Books.descr_gen(),
    'book_type':Books.list_pick(['Fiction', 'Non - Fiction'],False,[0.8,0.2]),
    'author_sex':Books.list_pick(['Male', 'Female'],False,[0.4,0.6]),
    'genre':Books.list_pick(['Satire', 'Drama','Romance', 'Self help'],False,[0.4,0.4,0.1,0.1]),
    'publisher' : Books.list_pick(['Hachette', 'HarperCollins','Macmillan', 'Penguin Random'])})
```
